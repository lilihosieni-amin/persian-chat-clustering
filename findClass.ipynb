```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, SpectralClustering
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import hdbscan
import time
```

```python
n_clusters = 41
random_state=42
```

Read embedding data


```python
import pandas as pd
import numpy as np
import gc

data = pd.read_parquet('/content/drive/MyDrive/Doukt/Goftino/NLP/colab/data/embeddings.parquet')
data = data.drop('original_index', axis=1)
data.head(4)
```

```python
# Clustering evaluation function
def evaluate_clustering(data, labels):
    """Compute evaluation metrics for clustering results"""

    # If there's only one cluster (or all noise), return minimal info
    if len(np.unique(labels)) <= 1:
        return {
            'silhouette_score': None,
            'num_clusters': len(np.unique(labels)),
            'cluster_sizes': np.bincount(labels + 1 if -1 in labels else labels).tolist()
        }

    # Compute silhouette score only for assigned (non-noise) samples
    valid_samples = labels != -1 if -1 in labels else np.ones(len(labels), dtype=bool)

    if sum(valid_samples) > 1 and len(np.unique(labels[valid_samples])) > 1:
        sil_score = silhouette_score(data[valid_samples], labels[valid_samples])
    else:
        sil_score = None

    return {
        'silhouette_score': sil_score,
        'num_clusters': len(np.unique(labels[labels != -1])) if -1 in labels else len(np.unique(labels)),
        'cluster_sizes': np.bincount(labels[labels != -1] if -1 in labels else labels).tolist()
    }
```

K-means

```python
print("Running K-Means...")
start_time = time.time()

# Initialize and fit KMeans with 63 clusters
kmeans = KMeans(n_clusters=63, random_state=random_state, n_init=10)
labels = kmeans.fit_predict(data)

# Save cluster labels to DataFrame
result_df = pd.DataFrame(labels, columns=['clusters'])

execution_time = time.time() - start_time
print(f"K-Means finished in {execution_time:.2f} seconds.")
result_df.to_csv('/content/drive/MyDrive/Doukt/Goftino/NLP/colab/question/Findcluster/kmeans63_result.csv')

# Evaluate clustering results
results = evaluate_clustering(data, labels)
results['execution_time'] = execution_time
results['inertia'] = kmeans.inertia_
print(results)

# # Optional: visualize the clustering result
# visualize_clusters(data, labels, "K-Means Clustering")
```

HBDSCAN

```python
import pandas as pd

data_reduced = pd.read_csv('/content/drive/MyDrive/Doukt/Goftino/NLP/colab/data/reduceEmbedd.csv')
data_reduced.head(3)
```

```python
import hdbscan
print("Running HDBSCAN...")
start_time = time.time()

# Initialize and fit HDBSCAN
hdb = hdbscan.HDBSCAN(min_cluster_size=100, min_samples=10, cluster_selection_method='leaf')
labels = hdb.fit_predict(data_reduced)

# Create a DataFrame for cluster labels
result_df = pd.DataFrame(labels, columns=['clusters'])

execution_time = time.time() - start_time
print(f"HDBSCAN completed in {execution_time:.2f} seconds.")

# Save results to CSV
result_df.to_csv('/content/drive/MyDrive/Doukt/Goftino/NLP/colab/question/hdb_result.csv', index=False)

# Evaluate clustering results
results = evaluate_clustering(data_reduced, labels)
results['execution_time'] = execution_time
results['noise_points'] = sum(labels == -1)
print(results)

# Optional: visualize clustering results
# visualize_clusters(data, labels, "HDBSCAN Clustering")
```
